## Parse Excel Sheets with AIMs Observations

```python
# parse the low dose AIM spreadsheet
sh = a.sheet_by_name(u"low dose")
lowdose = []

for row in range(sh.nrows):
    # update the current day
    if type(sh.cell(rowx=row, colx=0).value) == unicode and sh.cell(rowx=row, colx=0).value[0:3] == "DAY":
        (day, dose) = sh.cell(rowx=row, colx=0).value.split("_")
        print day, dose
       
    print sh.cell(rowx=row, colx=0).value, type(sh.cell(rowx=row, colx=0).value)
    if (type(sh.cell(rowx=row, colx=0).value) == unicode or type(sh.cell(rowx=row, colx=0).value) == unicode) and len(sh.cell(rowx=row, colx=0).value.split("-")) == 2:
        (mouse_type, mouse_id) = sh.cell(rowx=row, colx=0).value.split("-")
        #    print sh.cell(rowx=row, colx=0).value

        
    for col in range(1, sh.ncols):
        if type(sh.cell(rowx=0, colx=col).value) == unicode and sh.cell(rowx=0, colx=col).value.split(" ")[1] == "min":
            curtime = sh.cell(rowx=0, colx=col).value.split(" ")[0]

        if type(sh.cell(rowx=row, colx=0).value) == unicode and len(sh.cell(rowx=row, colx=0).value.split("-")) == 2:
            if len(sh.cell(rowx=1, colx=col).value) > 0:
                measurement_type = sh.cell(rowx=1, colx=col).value
                val = sh.cell(rowx=row, colx=col).value
                if val != "":
                    lowdose.append( { "day" : day[3], "measurement_time" : curtime, 
                                     "dose" : dose, "mouse_type" : mouse_type, "mouse_id" : mouse_id, 
                                      "measurement_type" : measurement_type, "val" : val } )
        
        
```

```python
# parse the high dose AIM spreadsheet
sh = a.sheet_by_name(u"high dose")
highdose = []

for row in range(sh.nrows):
    # update the current day
    if type(sh.cell(rowx=row, colx=0).value) == unicode and sh.cell(rowx=row, colx=0).value[0:3] == "DAY":
        (day, dose) = sh.cell(rowx=row, colx=0).value.split("_")
#        print day, dose
       
    print sh.cell(rowx=row, colx=0).value, type(sh.cell(rowx=row, colx=0).value)
    if (type(sh.cell(rowx=row, colx=0).value) == unicode or type(sh.cell(rowx=row, colx=0).value) == unicode) and len(sh.cell(rowx=row, colx=0).value.split("-")) == 2:
        (mouse_type, mouse_id) = sh.cell(rowx=row, colx=0).value.split("-")
        #    print sh.cell(rowx=row, colx=0).value

        
    for col in range(1, sh.ncols):
        if type(sh.cell(rowx=0, colx=col).value) == unicode and sh.cell(rowx=0, colx=col).value.split(" ")[1] == "min":
            curtime = sh.cell(rowx=0, colx=col).value.split(" ")[0]

        if type(sh.cell(rowx=row, colx=0).value) == unicode and len(sh.cell(rowx=row, colx=0).value.split("-")) == 2:
            if len(sh.cell(rowx=1, colx=col).value) > 0:
                measurement_type = sh.cell(rowx=1, colx=col).value
                val = sh.cell(rowx=row, colx=col).value
                if val != "":
                    highdose.append( { "day" : day[3], "measurement_time" : curtime, 
                                     "dose" : dose, "mouse_type" : mouse_type, "mouse_id" : mouse_id, 
                                      "measurement_type" : measurement_type, "val" : val } )
                else:
                    highdose.append( { "day" : day[3], "measurement_time" : curtime, 
                                     "dose" : dose, "mouse_type" : mouse_type, "mouse_id" : mouse_id, 
                                      "measurement_type" : measurement_type, "val" : 0 } )
                    
highdose = pd.DataFrame(highdose)
```

## high dose, cp73

```python
def plot_aim_data(aim_data, dosegroup_name, cptype):    
    # get the sort order for all the individuals by the total sum, so that 
    # we can keep the order / colors consistent on all the plots
    series = (aim_data.select(lambda x: aim_data.mouse_type[x] == cptype)
                    .groupby(["mouse_type", "mouse_id","day"])
                    .sum().unstack().transpose())
    
    # get grand totals and sort individuals in that order
    z = series.sum()
    o = z.order(ascending=False)
    
    fig_high_cp73_aimcomponents = figure() 
    
    fig_high_cp73_aimcomponents.set_size_inches(25, 6)
    fig_high_cp73_aimcomponents.suptitle("AIM subcomponents vs. time by individual mouse - %(dosegroup_name)s, %(cptype)s" 
                                            % { "dosegroup_name" : dosegroup_name, "cptype" : cptype } )
    for (i, mtype) in enumerate(aim_data.measurement_type.unique()):
        ax = fig_high_cp73_aimcomponents.add_subplot(1,5,i+1)
        ax.set_axis_bgcolor("#F2F2F2")
        ax.set_color_cycle([get_cmap("Paired")(k) for k in linspace(0, 1, 12)])
        title(mtype)
        d = aim_data.select(lambda x: aim_data.measurement_type[x] == mtype and aim_data.mouse_type[x] == cptype)
        grouped = d.groupby(["mouse_type", "mouse_id"])
        for g in o.index:
            daygroups = grouped.get_group(g).groupby("day")
            ax.plot(daygroups.sum(), ".-", label=g[0] + " #" + g[1], ms=17, lw=2.5)
            ax.set_ylim((-1, 22))
            ax.set_xticks(range(len(daygroups.sum().index))) 
            ax.set_xticklabels(daygroups.sum().index)
            ax.set_ylabel("AIM score - sum over the 6 measurements")
            ax.set_xlabel("day")
        grid()
    l = ax.legend(bbox_to_anchor=(1.05, 0), loc='lower left', borderaxespad=0., title="individual mouse")        
    l.legendPatch.set_fc("#F2F2F2")
    
    
    fig_high_cp73_summary = figure()
    ax = fig_high_cp73_summary.add_subplot(1,2,1)
    title("Composite (summed) LiLoAxOl AIM score \n by individual by day, %(dosegroup_name)s, %(cptype)s" 
            % { "dosegroup_name" : dosegroup_name, "cptype" : cptype } )
    ax.set_color_cycle([get_cmap("Paired")(k) for k in linspace(0, 1, 12)])
    ax.set_ylim(-5, 60)
    ax.set_xlim(-0.5, 4.5)
    ax.set_axis_bgcolor("#F2F2F2")
    series = (aim_data.select(lambda x: aim_data.mouse_type[x] == cptype)
                    .groupby(["mouse_type", "mouse_id","day"])
                    .sum().unstack().transpose())
    lines = ax.plot( series[o.index],  "o-", ms=12, lw=2.5)
    ax.set_xticks(range(len(series.ix["val"].index)))
    ax.set_xticklabels(series.ix["val"].index)
    ax.set_xlabel("day")
    ax.set_ylabel("composite AIM score")
    grid()
    
    ax = fig_high_cp73_summary.add_subplot(1,6,4)
    title("Sum (over time) of composite LiLoAxOl AIM score by individual, \n %(dosegroup_name)s, %(cptype)s"
            % { "dosegroup_name" : dosegroup_name, "cptype" : cptype } )
    ax.set_xlim(-0.5, 0.5)
    ax.set_ylim(-5, 200)
    ax.set_axis_bgcolor("#F2F2F2")
    ax.set_color_cycle([get_cmap("Paired")(k) for k in linspace(0, 1, 12)])
    for v in o:
        ax.plot(v, "s", ms=10, alpha=0.7)
    grid()
    xlabel("")
    ax.set_ylabel("Total AIM score")
    ax.set_xticks([])
    ax.set_yticks(arange(0, 200, 10))
    l = ax.legend( lines, [g[0] + " #" + g[1] for g in o.index], bbox_to_anchor=(1.3, 0), loc='lower left', borderaxespad=0., title="individual mouse")
    l.legendPatch.set_fc("#F2F2F2")
    
    # li / ax / Ol
    fig_high_cp73_summary = figure()
    ax = fig_high_cp73_summary.add_subplot(1,2,1)
    title("Composite (summed) AIM LiAxOl score \n by individual by day, %(dosegroup_name)s, %(cptype)s"
            % { "dosegroup_name" : dosegroup_name, "cptype" : cptype } )
    ax.set_color_cycle([get_cmap("Paired")(k) for k in linspace(0, 1, 12)])
    ax.set_ylim(-5, 60)
    ax.set_xlim(-0.5, 4.5)
    ax.set_axis_bgcolor("#F2F2F2")
    series = (aim_data.select(lambda x: aim_data.mouse_type[x] == cptype and aim_data.measurement_type[x] in set(["Li", "Ax", "Ol"]))
                    .groupby(["mouse_type", "mouse_id","day"])
                    .sum().unstack().transpose())
    lines = ax.plot( series[o.index],  "o-", ms=12, lw=2.5)
    ax.set_xticks(range(len(series.ix["val"].index)))
    ax.set_xticklabels(series.ix["val"].index)
    ax.set_xlabel("day")
    ax.set_ylabel("composite AIM score")
    grid()
    
    ax = fig_high_cp73_summary.add_subplot(1,6,4)
    title("Sum (over time) of composite LiAxOl AIM score by individual, \n %(dosegroup_name)s, %(cptype)s"
            % { "dosegroup_name" : dosegroup_name, "cptype" : cptype } )
    ax.set_xlim(-0.5, 0.5)
    ax.set_ylim(-5, 175)
    ax.set_axis_bgcolor("#F2F2F2")
    ax.set_color_cycle([get_cmap("Paired")(k) for k in linspace(0, 1, 12)])
    for v in series.sum()[o.index]:
        ax.plot(v, "s", ms=10, alpha=0.7)
    grid()
    xlabel("")
    ax.set_ylabel("Total AIM score")
    ax.set_yticks(arange(0, 160, 10))
    ax.set_xticks([])
    l = ax.legend( lines, [g[0] + " #" + g[1] for g in o.index], bbox_to_anchor=(1.3, 0), loc='lower left', borderaxespad=0., title="individual mouse")
    l.legendPatch.set_fc("#F2F2F2")
```

```python
plot_aim_data( aim_data = lowdose, dosegroup_name = "Chronic Low Dose", cptype = "CP101")
plot_aim_data( aim_data = lowdose, dosegroup_name = "Chronic Low Dose", cptype = "CP73")
plot_aim_data( aim_data = highdose, dosegroup_name = "Chronic High Dose", cptype = "CP101")
plot_aim_data( aim_data = highdose, dosegroup_name = "Chronic High Dose", cptype = "CP73")
```
