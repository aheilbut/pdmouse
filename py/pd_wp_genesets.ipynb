```python
overlaps = alib.wikipath.get_overlaps(cs.symbol, wp.mouse_wp_symbols).sort("match_count", ascending=False)
```

```python
p = overlaps.apply(
 lambda x: (
  scipy.stats.hypergeom.sf(
    x.ix["match_count"]-1, # number of differentially expressed genes in set
    len(cp73_wide_info.symbol.unique()), # total number of genes
    x.ix["match_count"], # number of genes in current set
    len( cs.symbol.unique() ))), # total number of differentially expressed genes
 axis=1)
p = pandas.DataFrame(p, columns=["hypergeom p-val"])
```

```python
overlaps = overlaps.merge(p, left_index=True, right_index=True).sort("hypergeom p-val", ascending=True)
overlaps["bonferroni"] = statsmodels.sandbox.stats.multicomp.multipletests(overlaps.ix[:,"hypergeom p-val"], method="bonferroni")[1]
overlaps["b-h fdr adj pval"] = statsmodels.sandbox.stats.multicomp.multipletests(overlaps.ix[:,"hypergeom p-val"].fillna(1.0), method="fdr_bh")[1]
```

```python
overlaps.to_csv("/data/adrian/Dropbox/Projects/Broad/PD_mouse/results/jan30/high_vs_saline_wikipath_overlap.xls",  sep="\t")
```
